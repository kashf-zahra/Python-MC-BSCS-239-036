## 1️⃣ Strings

A **string** is text written inside quotes.  
Example: `'Hello'` or `"Python"`

We can join (combine) strings or use methods like `.upper()` and `.lower()`.


```python
# String examples
name = "Kashf"
greeting = "Hello"

print(greeting + " " + name)  # joining strings
print(name.upper())           # make all letters capital
print(name.lower())           # make all letters small

print("Python String")
print('Python "String"')
print("Python 'String'")
```

## 2️⃣ Boolean Values

Booleans represent **True** or **False**.  
They are mostly used in conditions or comparisons.


```python

a = 10
b = 5

print("a > b :", a > b)    # Greater than
print("a < b :", a < b)    # Less than
print("a == b:", a == b)   # Equal to
print("a != b:", a != b)   # Not equal to
print("a >= b:", a >= b)   # Greater than or equal to
print("a <= b:", a <= b)   # Less than or equal to
```

## 3️⃣ Taking Input from User

We can take input from the user using the `input()` function.


```python
name = input("Enter your name: ")
print("Hello", name + "!")

age = input("Enter your age: ")
print("You are", age, "years old.")
```
